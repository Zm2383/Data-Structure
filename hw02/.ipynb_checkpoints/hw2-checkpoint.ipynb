 # Selection of Data and Reshaping the Data
 

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Data/U.S._Chronic_Disease_Indicators__CDI_.csv')
```

```python
##1.Remove all columns you do not need for the analysis
data.info
```

```python
### Select columns necessary
columns_to_select = data[['YearStart','LocationDesc', 'LocationAbbr','Question','DataValueType','DataValue','Stratification1']]
```

```python
###Select binge drinking and Povery variables
data1= columns_to_select[columns_to_select['Question'].isin(['Binge drinking prevalence among adults aged >= 18 years','Poverty'])]
         
```

```python
###Select gender variable
data2 = data1[data1.Stratification1.isin(['Male', 'Female', 'Overall'])]
```

```python
###Select Crude Prevalence in Percent
data3=data2[data2['DataValueType'] == 'Crude Prevalence']
data3
```

## 2.Convert the dataset to a wide format data set using the commands from the pandas package.

```python
data3_wide=data3.pivot(index=['LocationDesc', 'LocationAbbr', 'YearStart'], columns= ['Stratification1', 'Question'], values='DataValue').reset_index()
data3_wide
```

## 3.Rename the variables 

```python
data4=data3_wide.copy()
data4.columns=['state', 'stateabb', 'year', 'poverty_female', 'poverty_male', 'poverty', 'binge_female', 'binge_male', 'binge_all']
data4
```

```python
data5=data4.drop(['poverty_female', 'poverty_male'],1)
data5[['poverty', 'binge_female', 'binge_male', 'binge_all']] = data5[['poverty', 'binge_female', 'binge_male', 'binge_all']].apply(pd.to_numeric)
```

# Data Transformation and Summary Results
## 4.Produce a table that shows the overall, female, and male binge drinking prevalences across U.S. States in the most recent year of data for the Top 10 binge drinking states

```python
data_select=data5[['state','year','binge_all','binge_female', 'binge_male']]
data_select_2019=data_select[data_select['year']==2019]
data_select_2019
```

```python
top10_data_2019 = data_select_2019.sort_values(by='binge_all', ascending=False)
top10_data_2019.head(10)
```

## 5.Calculate the average annual growth rates (in percent) of overall binge drinking across states for the years the data is available

```python
###Select the columns
data6=data5.copy()
data6['binge_all'] = data6['binge_all'].astype(float)
data6 = data6[['state','stateabb','binge_all']]
###Calculate the average annual growth rate
data6['changes'] = data6['binge_all'].pct_change()
avg_growth = pd.DataFrame(data6.groupby('state')['changes'].mean().reset_index(name = 'avg_growth'))
avg_growth
```

```python
topinc5 = avg_growth.sort_values('avg_growth',ascending = False).head(5)
topinc5['avg_growth']=topinc5['avg_growth'].apply(lambda x:'%.2f%%' % (x*100))
topinc5
```

```python
topdec5 = avg_growth.sort_values('avg_growth').head(5)
topdec5['avg_growth']=topdec5['avg_growth'].apply(lambda x:'%.2f%%' % (x*100))
topdec5
```
